```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Создаем список количества потоков
p_threads = np.arange(10) + 1

# Вводим скорости работы программы на фиксированном количесвте потоков
speed_1 = np.array([0.020, 0.020, 0.020, 0.020, 0.019, 0.019, 0.019, 0.019, 0.019, 0.018])
speed_2 = np.array([0.018, 0.018, 0.017, 0.017, 0.017, 0.017, 0.017, 0.017, 0.017, 0.017])
speed_3 = np.array([0.017, 0.017, 0.017, 0.017, 0.017, 0.016, 0.016, 0.016, 0.016, 0.016])
speed_4 = np.array([0.017, 0.017, 0.017, 0.017, 0.017, 0.017, 0.017, 0.017, 0.017, 0.016])
speed_5 = np.array([0.021, 0.019, 0.019, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018, 0.018])
speed_6 = np.array([0.021, 0.021, 0.020, 0.020, 0.020, 0.020, 0.020, 0.020, 0.020, 0.020])
speed_7 = np.array([0.025, 0.024, 0.024, 0.024, 0.024, 0.023, 0.023, 0.023, 0.023, 0.023])
speed_8 = np.array([0.027, 0.027, 0.027, 0.027, 0.026, 0.026, 0.026, 0.026, 0.025, 0.025])
speed_9 = np.array([0.031, 0.030, 0.030, 0.030, 0.030, 0.030, 0.030, 0.030, 0.029, 0.028])
speed_10 = np.array([0.033, 0.033, 0.033, 0.033, 0.033, 0.033, 0.033, 0.033, 0.032, 0.032])

# Усредняем скорости работы программы на фиксированном потоке
speeds = [speed_2, speed_3, speed_4, speed_5, speed_6, speed_7, speed_8, speed_9, speed_10]
common_speed = speed_1.copy()

for speed in speeds:
    common_speed = np.vstack((common_speed, speed))

common_speed = np.mean(common_speed, axis=1)

# Находим ускорение
acceleration = common_speed[0]/common_speed

# Находим эффективность
efficiency = acceleration/p_threads
```

```python
print(common_speed)
print('acceleration: ', acceleration)
print('efficiency: ', efficiency)
```

```python
# Строим график ускорения работы программы
#plt.figure(figsize=(15,10))

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15,5))


axs[0].plot(p_threads, acceleration,'-r', label='real acceleration')
axs[0].plot(p_threads,p_threads, '-b', label='ideal acceleration')

axs[1].plot(p_threads, efficiency,'-r', label='real eficiency')
axs[1].plot(p_threads,p_threads/p_threads, '-b', label='ideal efficiency')

axs[0].set_xlabel('Количество задействованных потоков')
axs[0].set_ylabel('Величина ускорения, S')

axs[1].set_xlabel('Количество задействованных потоков')
axs[1].set_ylabel('Величина эффективности, E')

axs[0].set_title('Ускорение работы программы, S')
axs[1].set_title('Эффективность работы программы, E')

axs[0].legend()
axs[1].legend()

axs[0].grid()
axs[1].grid()

axs[0].axis('equal')
axs[1].axis('equal')

plt.show()
```
