```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
docs = ['test_0'+str(i)+'.json' for i in range(10)]
docs = [pd.read_json(docs[i], encoding='utf-8') for i in range(10)]
docs
```

```python
df_docs = pd.concat(docs)
new_ind = np.arange(df_docs.size)
df_docs.index = new_ind

df_docs.head()
```

```python
tasks = []

df_docs_matrix = df_docs.values
for i in range(df_docs_matrix.shape[0]):
    if 'ГЛАВНАЯ' in df_docs_matrix[i][0]['text']:
        tasks.append(df_docs_matrix[i][0])
        
    elif 'главная' in df_docs_matrix[i][0]['text']:
        tasks.append(df_docs_matrix[i][0])
```

```python
task_id = []
text = []
ch_id = []
ch_text = []
target = []

count = 0

for i in tasks:
    count = len(i['question']['choices'])
    for j in i['question']['choices']:
        task_id.append(i['id'])
        text.append(i['text'])
        ch_id.append(j['id'])
        ch_text.append(j['text'])
        if str(ch_id[-1]) in i['solution']['correct_variants'][0]:
            target.append(1)
        else:
            target.append(0)
        
dataset = pd.DataFrame({'target' : target,
                        'task_id' : task_id, 
                        'text' : text, 
                        'ch_id' : ch_id, 
                        'ch_text' : ch_text})
dataset
```

```python
dataset['text_splits'] = dataset['text'].apply(lambda x: x.split())
dataset['ch_splits'] = dataset['ch_text'].apply(lambda x: x.split())

dataset = dataset.drop('text', axis=1)
dataset = dataset.drop('ch_text', axis=1)

dataset = dataset[['target', 'task_id', 'text_splits', 'ch_id', 'ch_splits']]
dataset['intersection'] = [len(set(a).intersection(b)) for a, b in zip(dataset.text_splits, dataset.ch_splits)]
dataset['diference'] = [len(set(a).difference(b)) for a, b in zip(dataset.ch_splits, dataset.text_splits)]
dataset
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
```

```python
X_train =  dataset.drop(['target', 'task_id', 'text_splits', 'ch_id', 'ch_splits'], axis = 1).iloc[:35, :]
Y_train = dataset['target'].iloc[:35]

X_test =  dataset.drop(['target', 'task_id', 'text_splits', 'ch_id', 'ch_splits'], axis = 1).iloc[35:, :]
Y_test = dataset['target'].iloc[35:]
```

```python
clf = RandomForestClassifier(random_state=24)
clf.fit(X_train, Y_train)
print('train:', roc_auc_score(Y_train, clf.predict(X_train)))
print('test:', roc_auc_score(Y_test, clf.predict(X_test)))
```

```python
import difflib as dl
```

```python
def string_differ(task, answer, ratio):
    d = dl.SequenceMatcher(None, task, answer)
    ratio = np.append(ratio, d.ratio())
    return ratio

def differ(tasks, answers):
    ratio = np.array([])
    for i in range(len(tasks)):
        ratio = string_differ(tasks[i], answers[i], ratio)
    return ratio
```

```python
ratio = differ(dataset.text_splits, dataset.ch_splits)
```

```python
DS = pd.DataFrame(data = {'target' : dataset.target,
                          'intersection' : dataset.intersection,
                          'diference' : dataset.diference,
                          'ratio' : ratio,
                          })
DS.head()
```

```python
X_train =  DS.drop(['target'], axis = 1).iloc[:35, :]
Y_train = DS['target'].iloc[:35]

X_test =  DS.drop(['target'], axis = 1).iloc[35:, :]
Y_test = DS['target'].iloc[35:]
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, Y_train)
print('train:', roc_auc_score(Y_train, clf.predict(X_train)))
print('test:', roc_auc_score(Y_test, clf.predict(X_test)))
```
